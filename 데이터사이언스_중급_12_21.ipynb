```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from umap import UMAP
```

```python
#Iris 데이터셋 로드
iris = datasets.load_iris()
X=iris.data
y=iris.target
target_names = iris.target_names
```

```python
# PCA 실행
pca = PCA(n_components=2)
X_r1=pca.fit_transform(X)
```

```python
#t-SNE 실행
tsne = TSNE(n_components=2, random_state=0)
X_r2=tsne.fit_transform(X)
```

```python
#UMAP 실행
umap = UMAP(n_components=2, random_state=0)
X_r3=umap.fit_transform(X)
```

```python
#결과 시각화
plt.figure(figsize=(15,5))

#pca결과
plt.subplot(1,3,1)
for color, i, target_name in zip(["navy", "turquoise", "darkorange"], [0,1,2], target_names):
  plt.scatter(X_r1[y == i,0], X_r1[y==i,1], color=color, alpha=.8,lw=2, label=target_name)
plt.title('PCA of IRIS dataset')
plt.xlabel('PC1 Componet')
plt.ylabel('PC2 Componet')

#t sne결과
plt.subplot(1,3,2)
for color, i, target_name in zip(["navy", "turquoise", "darkorange"], [0,1,2], target_names):
  plt.scatter(X_r2[y==i,0],X_r2[y==i,1], color=color, alpha=.8,lw=2, label=target_name)
plt.title('t-SNE of IRIS dataset')
plt.xlabel('t-SNE 1st Componet')
plt.ylabel('t-SNE 2st Componet')

#umap결과
plt.subplot(1,3,3)
for color, i, target_name in zip(["navy", "turquoise", "darkorange"], [0,1,2], target_names):
  plt.scatter(X_r3[y==i,0],X_r3[y==i,1], color=color, alpha=.8,lw=2, label=target_name)
plt.title('UMAP of IRIS dataset')
plt.xlabel('UMAP 1st Componet')
plt.ylabel('UMAP 2st Componet')
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from umap import UMAP
from sklearn.datasets import fetch_openml
```

```python
#MNIST 데이터넷 로드
mnist = fetch_openml('mnist_784', version=1)
X= mnist.data / 255.0 # 데이터 정규화
y=mnist.target
```

```python
# 시각화를 위해 데이터의 일부만
n_samples = 3000
X_subset=X[:n_samples]
y_subset=y[:n_samples]
```

```python
# PCA 실행
pca = PCA(n_components=2)
X_pca=pca.fit_transform(X_subset)

#t-SNE 실행
tsne = TSNE(n_components=2, random_state=0)
X_tsne=tsne.fit_transform(X_subset)

#UMAP 실행
umap = UMAP(n_components=2, random_state=0)
X_umap=umap.fit_transform(X_subset)
```

```python
# 결과 시각화
fig, axes = plt.subplots(1,3,figsize=(15,5))

# PCA 결과
axes[0].scatter(X_pca[:,0],X_pca[:,1], c=y_subset.astype(int), cmap='Spectral', s=5)
axes[0].set_title('PCA of MNIST dataset')

#t-SNE 결과
axes[1].scatter(X_tsne[:,0],X_tsne[:,1], c=y_subset.astype(int), cmap='Spectral', s=5)
axes[1].set_title('t-SNE of MNIST dataset')

#UMAP 결과
axes[2].scatter(X_umap[:,0],X_umap[:,1], c=y_subset.astype(int), cmap='Spectral', s=5)
axes[2].set_title('UMAP of MNIST dataset')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 데이터셋 불러오기
diabtes = load_diabetes()
X=diabtes.data
y=diabtes.target

#데이터를 훈련세트와 테스트 세트로 분할
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
# 모델의 생성 및 훈련
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
# 훈련 데이터에 대한 예측 수행
y_val = model.predict(X_train)

# 성능 지표 계산
mse_val = mean_squared_error(y_train,y_val)
r2_val = r2_score(y_train,y_val)

print(f"Mean Squared Error: {mse_val}")
print(f"R2 Score: {r2_val}")
```

```python
# 테스트 데이터에 대한 예측 수행
y_pred = model.predict(X_test)

# 성능 지표 계산
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
# 테스트 데이터에 대한 예측 수행
y_pred = model.predict(X_test)

#실제 값과 예측 값을 scatter plot으로 시각화
plt.scatter(y_test,y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('y_test vs. y_pred')
plt.show()
```

# KNN

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Iris 데이터 셋 로드
irsi = load_iris()
X=iris.data
y=iris.target

# 데이터셋을 훈련세트와 테스트 세트로 분할(2:1 비율)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=1/3, random_state=42)
```

```python
#KNN 모델 생성 및 훈련
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
# 테스트 데이터에 대한 예측
y_pred=knn.predict(X_test)

# 정확도 계산
accuracy = accuracy_score(y_test,y_pred)

print(f"Accuracy: {accuracy}")
```

#K-MEANS

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from sklearn.cluster import KMeans
#아이리스는 이미 했고
from sklearn.metrics import silhouette_score
```

```python
# load the Irsi dataset
iris = load_iris()
X=iris.data

#create a DataFrame for easier data manipulation
iris_df=pd.DataFrame(X,columns=iris.feature_names)
```

```python
#Initialize ~
silhouette_scores=[]

#plotting setup
plt.figure(figsize=(15,10))

#Loop through cluster sizes from 2 to 5
for i in range(2,6):
  #KMeans clustering
  kmeans = KMeans(n_clusters=i, random_state=42)
  labels = kmeans.fit_predict(X)

  #silhouette_score
  score = silhouette_score(X,labels)
  silhouette_scores.append(score)

  #scatter plot for each cluster result
  plt.subplot(2,2,i-1)
  sns.scatterplot(x=iris_df.iloc[:,0], y=iris_df.iloc[:,1], hue=labels, palette='bright')
  plt.legend(title='Cluster')
```
